###  Import the necessary libraries

```python
import pandas as pd
```

### Import the dataset from URL and assign it to a variable apple

```python
url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/09_Time_Series/Apple_Stock/appl_1980_2014.csv'
```

```python
apple=pd.read_csv(url)
```

```python
apple.head()
```

### Check out the type of the columns

```python
apple.dtypes
```

### Transform the Date column as a datetime type

```python
apple['Date']=pd.to_datetime(apple['Date'])
```

### Set the date as the index

```python
apple.set_index('Date',inplace=True)
```

### Is there any duplicate dates?

```python
apple[apple.index.duplicated()]
```

### Sort the index to be the first entry the oldest date

```python
apple=apple.sort_index(ascending=True)
apple.head()
```

### Get the last business day of each month

```python
apple.groupby(apple.index.strftime('%Y-%m')).tail(1)
```

### What is the difference in days between the first day and the oldest

```python
apple.index[-1]-apple.index[0]
```

```python
import numpy as np
```

### How many months in the data we have?

```python
apple.index[-1]
```

```python
apple.index[0]
```

```python
(apple.index[-1]-apple.index[0])/ np.timedelta64(1, 'M')
```

```python
print('{} mesecev'.format(int((apple.index[-1]-apple.index[0])/ np.timedelta64(1, 'M'))))
```
